```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from scipy import signal as sgn
import matplotlib
import matplotlib.pyplot as plt
```

```python
def set_ticks_array(minVal,maxVal,incVal):
    ticks_array=np.linspace(minVal,maxVal, int((maxVal-minVal)/incVal+1),endpoint=True)
    return ticks_array
```

```python
frekuensi=8
frekuensi_sampling=100
fasa_awal=0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A=1                                                        # amplitudo
B=0                                                        # offset
t = np.linspace(0,0.5,frekuensi_sampling)                  # x
s = B+A*np.sin(2*np.pi*frekuensi*t+np.deg2rad(fasa_awal))  # y



plt.stem(t,s)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('Sine Plot')
# plt.xlim(0,0.25)
# plt.ylim(-1.1,1.1)
# plt.yticks(set_ticks_array(-1,1,0.5))
# plt.xticks(set_ticks_array(0,0.25,0.05))
plt.grid()

plt.show()
```

Untuk membuat plot per bar ganti plt.plot() menjadi plt.stem()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stem.html

Untuk membuat sinyal segitiga dapat menggunakan signal.sawtooth()

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.sawtooth.html

Untuk membuat sinyal kotak dapat menggunakan signal.square()
https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.square.html

- **t** = np.linspace(0,duration,frekuensi_sampling) = awal dimulai dari 0 sampai 1.5 yaitu durasi waktunya dan jumlah titik sebanyak 100 titik

- **s** = B+Anp.sin(2np.pi*frekuensit *+np.deg2rad(fasa_awal)) = rumus gelombang sinusnya
- **s2** = B+A*np.sin(2*np.pi*4 *t+np.deg2rad(fasa_awal))= Ini gelombang sinus kedua dengan frekuensi 4 Hz.
