```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sgn
from IPython.display import Audio
```

## Pendahuluan
Berikut diberikan program sebagai berikut

```python
f=4                                                        # frekuensi
dur = 1                                                    # panjang/durasi dari sinyal (sepanjang sumbu x)
fs=100                                                   # frekuensi sampling
fasa_awal=0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A=1                                                        # amplitudo
B=0                                                        # offset
t = np.linspace(0,dur,fs*dur)                  # x
s = B+A*np.sin(2*np.pi*f*t+np.deg2rad(fasa_awal))  # y

plt.plot(t,s)
# plt.xlabel('Time (s)')
# plt.ylabel('Voltage (mV)')
# plt.title('Sine Plot')
# plt.xlim(0,0.25)
# plt.ylim(-1.1,1.1)
# plt.yticks(set_ticks_array(-1,1,0.5))
# plt.xticks(set_ticks_array(0,0.25,0.05))
plt.grid()

plt.show()
```

Dimana,

b. Ubah frekuensi sampling menjadi 8 Hz, apa yang terjadi?

```python
#Tulis Program disini

f = 4                                                        # frekuensi
dur = 1                                                      # panjang/durasi dari sinyal (sepanjang sumbu x)
fs = 8                                                       # frekuensi sampling
fasa_awal = 0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A = 1                                                        # amplitudo
B = 0                                                        # offset
t = np.linspace(0,dur,fs*dur)                                # x
s = B+A*np.sin(2*np.pi*f*t+np.deg2rad(fasa_awal))            # y

plt.plot(t,s)
# plt.xlabel('Time (s)')
# plt.ylabel('Voltage (mV)')
# plt.title('Sine Plot')
# plt.xlim(0,0.25)
# plt.ylim(-1.1,1.1)
# plt.yticks(set_ticks_array(-1,1,0.5))
# plt.xticks(set_ticks_array(0,0.25,0.05))
plt.grid()

plt.show()
```

Sehingga tampaknya plot/kurva yang dihasilkan terlihat patah-patah dan tidak lagi menyerupai gelombang sinus yang sempurna. Hal ini terjadi karena tidak cukupnya jumlah titik  untuk menangkap perubahan sinyal dengan baik.


c. Berapa frekuensi sampling minimal untuk mendapatkan jumlah gelombang yang sama dengan frekuensi-nya. Dan mengapa?


Jika secara teori frekuensi sampling minimalnya adalah 8Hz itu sudah cukup, karena nilai minimal frekuensi sampling itu 2 kalinya frekuensi tersebut ( 2 x f ). Namun, secara visual, yang terlihat hanya 3 gelombang, bukan 4 seperti yang diharapkan. yang berarti kalau ingin terlihat secara visual adanya 4 gelombang sesuai dengan frekuensi yaitu 4 Hz, maka frekuensi sampling dinaikan sedikit  misalnya 10 Hz. singkatnya dengan pemahaman berikut ini:

*  fs = 2f  -> jumlah gelombang benar, tapi bentuknya terlihat patah-patah / kasar.
*  fs > 2f  -> jumlah gelombang benar, dan bentuknya lebih halus.
*  fs < 2f  -> jumlah sampel tidak cukup untuk merepresentasikan sinyal dengan benar, sehingga bentuk gelombang yang terlihat bisa salah atau berbeda dari yang seharusnya.



d. Ubah frekuensi sampling menjadi 8 Hz, dan gunakan fungsi Cos sebagai fungsi sinyal

```python
#Tulis Program disini untuk membuktikan jumlah gelombangnya

#Tulis Program disini

f = 4                                                        # frekuensi
dur = 1                                                      # panjang/durasi dari sinyal (sepanjang sumbu x)
fs = 8                                                       # frekuensi sampling
fasa_awal = 0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A = 1                                                        # amplitudo
B = 0                                                        # offset
t = np.linspace(0,dur,fs*dur)                                # x
s = B+A*np.cos(2*np.pi*f*t+np.deg2rad(fasa_awal))            # y

plt.plot(t,s)
# plt.xlabel('Time (s)')
# plt.ylabel('Voltage (mV)')
# plt.title('Sine Plot')
# plt.xlim(0,0.25)
# plt.ylim(-1.1,1.1)
# plt.yticks(set_ticks_array(-1,1,0.5))
# plt.xticks(set_ticks_array(0,0.25,0.05))
plt.grid()

plt.show()
```
