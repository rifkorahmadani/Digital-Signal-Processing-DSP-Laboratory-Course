#Modul 5 - Konvolusi

Konvolusi adalah operasi matematika yang menggabungkan dua fungsi untuk menghasilkan fungsi baru.

$$y[n]=\sum_{k=0}^{M-1}h[k]\cdot x[n-k]$$

secara umum $x[k]$ didefinisikan sebagai sinyal input dan $h[k]$ adalah respon impuls dari sistem. Notasi konvolusi dapat ditulis sebagai berikut

$$y[n]=h[n]\ast x[n] $$

2. Buatlah program secara manual yang dapat mencari hasil konvolusi antara $x[n]$ dengan $h[n]$ lalu plot hasilnya dengan menggunakan plt.stem (jangan lupa import library nya). Beberapa fungsi yang mungkin dapat membantu,
- ```len()``` untuk menghitung jumlah/panjang elemen dalam array

*  Masukkan Data Sinyal

     Diketahui:

     x = [1, 3, 5, 2, 4]

     h = [1, 2, 3, 2, 4]

*  Hitung Panjang Sinyal Hasil (y[n])

   panjang hasil konvolusinya adalah: len_y = len(x) + len(h) - 1 = 5 + 5 - 1 = 9

*  Menginisialisasikan Array Hasil

   Buat array y[n] berisi nol sebanyak len_ y. [0, 0, 0, 0, 0, 0, 0, 0, 0], yang terdapat 9 elemen

*  Proses Konvolusi Secara Manual

   Konvolusi dilakukan dengan dua perulangan, di mana untuk setiap n dan k, jika n - k valid, maka x[k] * h[n - k] dihitung dan ditambahkan ke y[n].

*  Cetak Konvolusi hasil akhirnya

   Output: y[n] = [1, 5, 14, 23, 33, 36, 36, 16, 16]

*  Tammpilkan grafik

   Menggunakan plt.stem() agar lebih mudah melihat bentuk sinyal hasil

````python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from scipy import signal as sgn
import matplotlib
import matplotlib.pyplot as plt
````

````python
def set_ticks_array(minVal,maxVal,incVal):
    ticks_array=np.linspace(minVal,maxVal, int((maxVal-minVal)/incVal+1),endpoint=True)
    return ticks_array
````

````python
x = [1, 3, 5, 2, 4]
h = [1, 2, 3, 2, 4]

len_y = len(x) + len(h) - 1
y = np.zeros(len_y)

for n in range(len_y):
    for k in range(len(x)):
        if 0 <= n - k < len(h):
            y[n] += x[k] * h[n - k]

print("y[n] =", y.astype(int))
n = np.linspace(0,8,9, endpoint=True)
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Hasil Konvolusi Manual antara h[n] dan x[n]')
plt.grid()
plt.show()
````

6. Jika ```h[n]=[0 0 0 0 0 0 1 0]``` lakukan konvolusi dengan sinyal ```x[n]``` dan analisis hasilnya (baik dengan fungsi manual maupun dengan np.convolve)

````python
#Tuliskan program disini (fungsi manual)

x = [1, 3, 5, 2, 4]
h = [0, 0, 0, 0, 0, 0, 1, 0]

len_y = len(x) + len(h) - 1
y = np.zeros(len_y)

for n in range(len_y):
    for k in range(len(x)):
        if 0 <= n - k < len(h):
            y[n] += x[k] * h[n - k]

print("y[n] =", y.astype(int))
n = np.linspace(0,11,12, endpoint=True)
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Hasil Konvolusi Manual antara h[n] dan x[n]')
plt.grid()
plt.show()
````

7. Jika ```x[n] = [1 3 1 5 3 1]``` tentukan ```h[n]``` agar amplitudo dari hasil konvolusinya menunjukkan selisih antar sinyal ```x[n]``` (hasil konvolusinya ialah [1 2 -2 4 -2 -2 1])

````python
#Tuliskan program disini

x = [1, 3, 1, 5, 3, 1]
h = [1, -1]

len_y = len(x) + len(h) - 1
y = np.zeros(len_y)

for n in range(len_y):
    for k in range(len(x)):
        if 0 <= n - k < len(h):
            y[n] += x[k] * h[n - k]

print("y[n] =", y.astype(int))
n = np.linspace(0,6,7, endpoint=True)
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Hasil Konvolusi Manual antara h[n] dan x[n]')
plt.grid()
plt.show()
````

Dari program tersebut untuk mencari nilai h[n], berikut ini

Diketahui:

x[n] =  [1 3 1 5 3 1]

y[n] = [1 2 -2 4 -2 -2 1]

jika,

 h[n] = [1,-1]

maka:

h[0] * x[n] = 1 3 1 5 3 1
