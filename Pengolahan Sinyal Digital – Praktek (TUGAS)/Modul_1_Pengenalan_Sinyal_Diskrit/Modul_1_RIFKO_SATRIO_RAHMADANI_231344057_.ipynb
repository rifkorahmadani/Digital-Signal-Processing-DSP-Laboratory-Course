```python
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
print(f"Version of Numpy:      {np.__version__}")
print(f"Version of Matplotlib: {matplotlib.__version__}")
import math
from matplotlib import animation
```

## Generasi Sinyal Periodik Sederhana

```python
def set_ticks_array(minVal,maxVal,incVal):
    ticks_array=np.linspace(minVal,maxVal, int((maxVal-minVal)/incVal+1),endpoint=True)
    return ticks_array
```

```python
frekuensi=12
frekuensi_sampling=100
fasa_awal=0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A=1                                                        # amplitudo
B=0                                                        # offset

t = np.linspace(0,0.25,frekuensi_sampling)                 # x
s = B+A*np.sin(2*np.pi*frekuensi*t+np.deg2rad(fasa_awal))  # y


plt.plot(t,s)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('Sine Plot')
plt.xlim(0,0.25)
plt.ylim(-1,1)
plt.yticks(set_ticks_array(-1,1,0.5))
plt.xticks(set_ticks_array(0,0.25,0.05))
plt.grid()

plt.show()

```

```python
#Tuliskan Program Anda disini

frekuensi=1/26.6
frekuensi_sampling=100
fasa_awal=0                                                # fasa awal, jika ditulis sebagai derajat maka perlu dikonversikan ke radian
A=2                                                        # amplitudo
B=0                                                        # offset

t = np.linspace(0,52.5,frekuensi_sampling)                 # x
s = B+A*np.sin(2*np.pi*frekuensi*t+np.deg2rad(fasa_awal))  # y

t1 = np.linspace(0,52.5,frekuensi_sampling)                 # x
s1 = B+A*np.cos(2*np.pi*frekuensi*t+np.deg2rad(fasa_awal))

plt.plot(t,s)
plt.plot(t1,s1)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('Cos (Blue)/Sin (Red) Plot (Note Phase Difference)')
plt.xlim(0,52.5)
plt.ylim(-1,1)
plt.yticks(set_ticks_array(-2,2,0.5))
plt.xticks(set_ticks_array(0,60,10))
plt.grid()

plt.show()
```

> *Jelaskan Program Anda Disini*

Pada soal ini terdapat 2 gelombang yang dapat diketahui yaitu gelombang sinus (warna biru) dan gelombang cosinus (warna orange). programnya hampir sama dengan soal no. 3 namun pada program ini ditambahkan saja untuk gelombang cosinusnya, yaitu:
