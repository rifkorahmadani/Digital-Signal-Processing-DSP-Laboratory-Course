#Modul 6 - Konvolusi

Konvolusi adalah operasi matematika yang menggabungkan dua fungsi untuk menghasilkan fungsi baru.

$$y[n]=\sum_{k=0}^{M-1}h[k]\cdot x[n-k]$$

secara umum $x[k]$ didefinisikan sebagai sinyal input dan $h[k]$ adalah respon impuls dari sistem. Notasi konvolusi dapat ditulis sebagai berikut

$$y[n]=h[n]\ast x[n] $$

## Soal

### 1. Bila diketahui suatu output suatu sistem ialah $$y[n]=-x[n] +  x[n-1] $$

a. Tuliskan fungsi $h[n] = $...



fungsi nya adalah , **$$h[n] = -δ[n]+δ[n-1]$$**

$$h[0].-δ[n] → -δ[n] → h[0]=-1$$
$$h[1].δ[n-1] → δ[n-1] → h[1]=1$$

mengapa? karena dengan mecocokkan dalam bentuk persamaan fungsi tersebut sebenarnya hampir sama dengan fungsi sistem output tersebut, hanya saja fungsi h[n] ini menjadi impuls respon sistem.

b. Buat konvolusi $$y[n]=h[n]\ast x[n] $$ untuk $x[n]=[3, 4, 2, 2, 2, 1, 5, 3]$

````python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from scipy import signal as sgn
import matplotlib
import matplotlib.pyplot as plt
````

````python
def set_ticks_array(minVal,maxVal,incVal):
    ticks_array=np.linspace(minVal,maxVal, int((maxVal-minVal)/incVal+1),endpoint=True)
    return ticks_array
````

````python
#Tuliskan program disini plot input
x = [3, 4, 2, 2, 2, 1, 5, 3]
h = [-1, 1]

plt.stem(x)
plt.title("Grafik untuk x[n]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()
````

````python
#Tuliskan program disini plot output
x = [3, 4, 2, 2, 2, 1, 5, 3]
h = [-1, 1]

output = np.convolve(x, h)
print("y[n] =", output)

plt.stem(output)
plt.title("Grafik Konvolusi untuk y[n] = h[n] * x[n]")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
````

c. Amati hasilnya, dan berikan penjelasan apa yang dilakukan oleh fungsi h[n]

**$$h[n] = -δ[n]+δ[n-1]$$**
fungsi h[n] tersebut berarti sedang mencari selisih antara dua nilai yang berdekatan pada sinyal x[n]. jadi. konvolusi ini membuat supaya bisa melihatatau mengetahui seberapa besar perubahan dari satu titik ke titik berikutnya. Simpelnya seperti memberi tahu seberapa besar dan ke arah mana peubahan terjadi antar nilai-nilai dalam sinyal

### 2. Bila diketahui suatu output suatu sistem ialah $$y[n]=x[n] -  2x[n-1]+x[n-2] $$

a. Tuliskan fungsi $h[n] = $...

fungsinya ialah, **$$h[n]=δ[n] -  2δ[n-1]+δ[n-2] $$**


$$h[0].δ[n] → δ[n] → h[0]= 1 $$
$$h[1].-2δ[n-1] → -2δ[n-1] → h[1]= -2$$
$$h[2].δ[n-2] → δ[n-2] → h[2]= 1$$
mengapa? karena dengan mecocokkan dalam bentuk persamaan fungsi tersebut sebenarnya hampir sama dengan fungsi sistem output tersebut, hanya saja fungsi h[n] ini menjadi impuls respon sistem.

b. Buat konvolusi $$y[n]=h[n]\ast x[n] $$ untuk $x[n]=[1, 2, 3, 5, 5, 5, 3, 2, 1]$ menggunakan ```np.convolve()``` tambahkan argumen ```mode='valid'``` didalam fungsi tersebut dan amati hasilnya

````python
#buat program disini plot input
x = [1,2,3,5,5,5,3,2,1]
h = [1,-2,1]
plt.stem(x)
plt.title("Grafuk untuk x[n]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()
````

````python
#buat program disini plot output
x = [1,2,3,5,5,5,3,2,1]
h = [1,-2,1]

output = np.convolve(x, h,  mode='valid')
print("y[n] =", output)

plt.stem(output)
plt.title("Grafik Konvolusi untuk y[n] = h[n] * x[n]")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
````

c. Amati hasilnya, dan berikan penjelasan apa yang dilakukan oleh fungsi h[n]

Menurut saya h[n] itu bisa dianggap sebagai filter, akan bernilai nol jika sinyalnya itu tetap, karena tidak ada perubahan arah. Tapi kalau sinyal berubah arah, misalnya dari naik-turun, maka hasil konvolusinya akan bernilai positif atau negatif. Ini karena filter tersebut mendeteksi perubahan bentuk sinyal, bukan hanya naik atau turunnya saja.

Selain itu, kalau kita pakai mode='valid', hasil konvolusi jadi lebih pendek karena hanya mengambil bagian yang sepenuhnya bisa diproses oleh filter, tidak seperti kalau tidak pakai mode='valid' akan menghitung semuanya termasuk pinggirannya.

### 3. Bila diketahui suatu output dari sistem ialah $$y[n]=1/3 ( x[n-2] +  x[n-3] +x[n-4])$$

a. Tuliskan fungsi $h[n] = $...

fungsi h[n] yaitu, **$$h[n]=1/3 (δ[n-2] +  δ[n-3] +δ[n-4]) $$**
 **$$h[n]=1/3δ[n-2] + 1/3 δ[n-3] + 1/3δ[n-4] $$**
$$h[0] = h[1] = 0$$
$$h[2].1/3 δ[n-2] → 1/3δ[n-2] → h[2] = 1/3 $$
$$h[3].1/3 δ[n-3] → 1/3δ[n-3] → h[3] = 1/3 $$
$$h[4].1/3 δ[n-4] → 1/3δ[n-4] → h[4] = 1/3 $$

mengapa? karena dengan mecocokkan dalam bentuk persamaan fungsi tersebut sebenarnya hampir sama dengan fungsi sistem output tersebut, hanya saja fungsi h[n] ini menjadi impuls respon sistem. Dan h[0] juga h[1] di nol kan karena memang tidak ada.


b. Buat konvolusi $$y[n]=h[n]\ast x[n] $$ untuk $x[n]=[1, 7, 3, 9, 14, 1, 7, 1, 14, 3, 9 , 2, 12]$ menggunakan ```np.convolve()``` tambahkan argumen ```mode='same'``` didalam fungsi tersebut dan amati hasilnya

````python
#buat program disini plot input
x = [1,7,3,9,14,1,7,1,14,3,9,2,12]
h = [0,0,1/3,1/3,1/3]

plt.stem(x)
plt.title("Grafuk untuk x[n]")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid()
plt.show()
````

````python
#Tuliskan program disini plot output
x = [1,7,3,9,14,1,7,1,14,3,9,2,12]
h = [0,0,1/3,1/3,1/3]

output = np.convolve(x, h, mode='same')
print("y[n] =", output)

plt.stem(output)
plt.title("Grafik Konvolusi untuk y[n] = h[n] * x[n]")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid()
plt.show()
````

> *Jelaskan argumen Anda untuk mendapatkan h[n]*

Saya memdapatkan h[n]  berdasarkan bentuk rumus output yaitu, $$y[n]=1/3 ( x[n-2] +  x[n-3] +x[n-4])$$

Dari rumus ini, terlihat bahwa output dihasilkan dari menjumlahkan tiga nilai input yang letaknya mundur (yaitu input ke-2, ke-3, dan ke-4 sebelumnya), lalu dibagi tiga. Ini artinya sistem hanya menggunakan tiga nilai input tersebut. Maka, saya buat h[n] berisi tiga angka 1/3 yang diletakkan di posisi ke-2, ke-3, dan ke-4. Fungsi  ini berperan untuk mengambil rata-rata dari tiga input sebelumnya. Jadi, setiap kali sistem bekerja, dia akan melihat ke belakang tiga langkah dan menghitung rata-ratanya. Karena itu, saya memilih h[n] = [0,0,1/3,1/3,1/3].
